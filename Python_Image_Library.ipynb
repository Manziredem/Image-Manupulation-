```python
# PIL is an image proccessing Library in Python. It allows you to manupulate Images

from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
# Redeading Image

image1 = Image.open("/content/drive/MyDrive/Colab Notebooks/fruits.jpeg")
image2 = Image.open("/content/drive/MyDrive/Colab Notebooks/tree.jpeg")
```

```python
# Display the 2nd image loaded

plt.imshow(image1)
plt.title("image 1 display")
plt.show()
```

```python
# Display the 2nd image loaded

plt.imshow(image2)
plt.title("Image 2 displayed")
plt.show()
```

```python
# properties of an image

print(image1.size) # the dimansion in terms of width and height 
print(image1.format)
print(image1.mode)
```

```python
# save the image

image1.save("newimg.png")
```

```python
# crop image

left = 50
top = 120
right = 250
bottom = 230
cropImage = image2.crop((left,top,right,bottom))
plt.imshow(cropImage)
plt.title("Croped image")
plt.show()
```

```python
# copy of image 

copiedImage = image1.copy()
plt.imshow(copiedImage)
plt.title("Copied image")
plt.show()
```

```python
# transposing: change the oriantation of an image e.g.: flip from right to left

tImage1 = image1.transpose(Image.FLIP_LEFT_RIGHT)
tImage2 = image1.transpose(Image.FLIP_TOP_BOTTOM)
tImage3 = image1.transpose(Image.ROTATE_90)
tImage4 = image1.transpose(Image.ROTATE_180)
tImage5 = image1.transpose(Image.ROTATE_270)
tImage6 = image1.transpose(Image.TRANSPOSE) #rows become columns and columns become rows 

plt.figure(figsize = (10,10))

plt.subplot(3,2,1)
plt.title('FLIP_LFT_RIGHT')
plt.imshow(tImage1)

plt.subplot(3,2,2)
plt.title('FLIP_TOP_BOTTOM')
plt.imshow(tImage2)

plt.subplot(3,2,3)
plt.title('ROTATE 90')
plt.imshow(tImage3)

plt.subplot(3,2,4)
plt.title('ROTATE 180')
plt.imshow(tImage4)

plt.subplot(3,2,5)
plt.title('ROTATE 270')
plt.imshow(tImage5)

plt.subplot(3,2,6)
plt.title('TRANSPOSE')
plt.imshow(tImage6)

plt.show()
```

```python
# resizing: the size are represented in aa taple from (wight, height). 
# They are different algorithm to do that: NEAREST, BOX, BIRIBEAR, HAMMING, and BICUBIC and LANZOS are quit better in terms of image resolution

# BILINEAR: for resize calculate the output pixel value using linear interpolation on all pixels that may contribute to the output value
# NEAREST: pich one nearest pixel from the input image. Poor result but simple
# BOX: each pixel source image contribute to one pixel of the dest. image with identical weights
# HAMMING: priduces a sharper image than BILINEAR, doen't have dislocations on local level with BOX
# BICUBIC: one of the better ones but expensive in terms of time
# LANCHZO: one of the better ones but expensive in terms of time

newsize = (100, 100)

plt.figure(figsize = (10,10))

resizedImage1 = image1.resize(newsize, Image.BILINEAR)
resizedImage2 = image1.resize(newsize, Image.NEAREST)
resizedImage3 = image1.resize(newsize, Image.BOX)
resizedImage4 = image1.resize(newsize, Image.HAMMING)
resizedImage5 = image1.resize(newsize, Image.BICUBIC)
resizedImage6 = image1.resize(newsize, Image.LANCZOS)

plt.subplot(3,2,1)
plt.title("BILINEAR")
plt.imshow(resizedImage1)

plt.subplot(3,2,2)
plt.title("NEAREST")
plt.imshow(resizedImage2)

plt.subplot(3,2,3)
plt.title("BOX")
plt.imshow(resizedImage3)

plt.subplot(3,2,4)
plt.title("HAMMING")
plt.imshow(resizedImage4)

plt.subplot(3,2,5)
plt.title("BICUBIC")
plt.imshow(resizedImage5)

plt.subplot(3,2,6)
plt.title("LANCZOS")
plt.imshow(resizedImage6)

plt.show()
```

```python
# Rotating

plt.figure(figsize = (10,10))

plt.subplot(3,2,1)
plt.title('Rotate 30')
angle = 30
rotatedImage = image1.rotate(angle)
plt.imshow(rotatedImage)

plt.subplot(3,2,2)
plt.title('Rotate -30')
angle = -30
rotatedImage = image1.rotate(angle)
plt.imshow(rotatedImage)

plt.show()
```

```python
# Convert an image into Numpy array

numpy_array = np.array(image1)
print(numpy_array) #this will output the tensor(set of matrices) by which this image is formed 
print(numpy_array.shape) # print the size shape of the array
```

```python
# If you have pixel array, you also convert it into an image

numpy_image = Image.fromarray(numpy_array)
plt.imshow(numpy_image)
plt.title("Image ")
plt.show()
```

```python
# Image Enhancement: changing some properties of an image like color, sharpness, contast, etc
# note: the default enhancement valur 1.0
from PIL import ImageEnhance

plt.figure(figsize = (10,10))

image_enh = image1.copy()

image1 = ImageEnhance.Color(image_enh).enhance(2.5)
# image2 = ImageEnhance.Cantrast(image_enh).enhance(2.5)
image3 = ImageEnhance.Brightness(image_enh).enhance(2.5)
image4 = ImageEnhance.Sharpness(image_enh).enhance(2.5)

plt.subplot(2,2,1)
plt.title("Color")
plt.imshow(image1)

plt.subplot(2,2,2)
plt.title("Cantrast")
plt.imshow(image2)

plt.subplot(2,2,3)
plt.title("Brightness")
plt.imshow(image3)

plt.subplot(2,2,4)
plt.title("Sharpness")
plt.imshow(image4)

plt.show()
```

```python
# Alpha Blending: mixing of two images. 
# formula: out image * (1.0 - alpha) + image2 * alpha
# Note: if the alpha = 0, you get the copy of the 1st image, when alpha = 1, you get the copy of 2nd image
# In the middle where alpha = 0.5 you the blending of both images
#*Note:* Before blending, make sure that the images have the same size and have alpha channel(usually .png has alpha chanel)

image_1 = image1.copy()
image_2 = image2.copy()

# Make both images the same size
image_2 = image_2.resize(image1.size)

#Blending
image_blend = Image.blend(image_1, image_2, .5)
plt.title("Two imaged blending")
plt.imshow(image_blend)

plt.show()
```

```python
# Image transform 

# the following is the affine transform --> the math behind it to be cohecked later om

image_transform = image1.copy()
image_transform = image_transform.transform(image_transform.size, Image.AFFINE,(1,-0.5,0.5 * image_transform.size[0],0,1,0))
plt.title("Transformed image - AFFINE")
plt.imshow(image_transform)
plt.show()
```

```python
# Crop out the image and zoom the croped part into the original size of the whole image

image_transform = image1.copy()
image_transform = image_transform.transform(image_transform.size, Image.EXTENT,(100,100, image_transform.size[0], image_transform.size[1]//3))
plt.title("Croped part zoomed")
plt.imshow(image_transform)
plt.show()
```

```python
# Flipping channel

# This is smething that is mostly used when using Tensorflow or pytoch when you are reverse RGB to like BGR

image_channels = image1.copy()
r,g,b = image_channels.split() # this will split the 3 channels into red, green and bleu
im1 = Image.merge("RGB", (b, g, r)) # marge them in them back in the oppisite order
im2 = Image.merge("RGB", (g, b, r))
im3 = Image.merge("RGB", (g, r, b))
im4 = Image.merge("RGB", (r, b, g))


plt.figure(figsize = (10,10))

plt.subplot(3,2,1)
plt.imshow(im1)
plt.title("Flipped from RGB to BGR")

plt.subplot(3,2,2)
plt.imshow(im2)
plt.title("Flipped from RGB to GBR")

plt.subplot(3,2,3)
plt.imshow(im3)
plt.title("Flipped from RGB to GRB")

plt.subplot(3,2,4)
plt.imshow(im4)
plt.title("Flipped from RGB to RBG")

plt.show()
```
